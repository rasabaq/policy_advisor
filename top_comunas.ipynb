# Top Comunas by Economic Complexity Index (ECI)
This notebook analyzes the `complexity_results.csv` to find the most complex comunas.

```python
import pandas as pd

# Load the complexity results
df = pd.read_csv('complexity_results.csv')

# Since ECI is a location-level metric, it is repeated for each ciuo (product) within the same comuna.
# We drop duplicates to get the unique ECI per comuna.
comunas_eci = df[['comuna', 'eci']].drop_duplicates()

# Sort by ECI descending to get the top comunas
top_comunas = comunas_eci.sort_values(by='eci', ascending=False)

print("Top 20 Comunas by Economic Complexity Index (ECI):\n")
display(top_comunas.head(20))
```
